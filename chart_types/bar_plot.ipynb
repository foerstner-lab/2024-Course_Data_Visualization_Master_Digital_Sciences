# Barplots 

## With matplotlib

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 30, 15, 20, 10]

# Creating the bar plot
plt.bar(categories, values)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot Example')

# Displaying the plot
plt.show()
```

## With seaborn

```python
import seaborn as sns

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 30, 15, 20, 10]

# Creating the bar plot using Seaborn
sns.barplot(x=categories, y=values)

# Adding labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Simple Bar Plot Example')

# Displaying the plot
plt.show()
```

## With bokeh

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 30, 15, 20, 10]

# Creating a ColumnDataSource
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Creating the figure
p = figure(x_range=categories, plot_height=350, title="Simple Bar Plot Example",
           toolbar_location=None, tools="")

# Creating the bars
p.vbar(x='categories', top='values', width=0.9, source=source)

# Adding labels and title
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Displaying the plot in the notebook
output_notebook()
show(p)
```
